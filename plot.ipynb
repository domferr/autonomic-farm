```python
import pandas as pd
import matplotlib.pyplot as plt
from glob import glob
```

```python
throughput_filename = glob("csv/throughput*.csv")[0];
df_throughput = pd.read_csv(throughput_filename)
df_throughput.describe()
```

```python
df_throughput.head()
```

```python
fig, axs = plt.subplots(figsize=(18, 4),dpi=300)
df_throughput.plot(ax=axs, x="time", y="throughput", color="black", title="Throughput over time")
axs.set_xlabel("time (usec)")
axs.set_ylabel("throughput (tasks/msecs)")

plt.show()
```
